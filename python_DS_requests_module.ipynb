```python
#@title JSON: Getting the JSON data and saving as CSV
# https://jsonplaceholder.typicode.com/
import requests
import csv

response = requests.get("https://jsonplaceholder.typicode.com/posts")
data = response.json()

print("Raw JSON data: ", data)

# Write to CSV
with open("posts.csv", mode="w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["User ID", "ID", "Title", "Body"])
    for post in data:
        writer.writerow([post["userId"], post["id"], post["title"], post["body"]])

print("Data saved to posts.csv")
```

```python
#@title JSON: Website Status Checker
import requests

websites = ["https://www.google.com", "https://www.github.com", "https://www.example.com", "http://wwwfake.123.com"]

for site in websites:
    try:
        response = requests.get(site)
        if response.status_code == 200:
            print(f"{site} is UP")
        else:
            print(f"{site} might be DOWN. Status code: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"{site} is DOWN. Error: {e}")
```

```python
#@title JSON: Translating Text (unicode)
import requests

def translate_text(text, source_lang, target_lang):
    url = "https://api.mymemory.translated.net/get"
    params = {
        "q": text,
        "langpair": f"{source_lang}|{target_lang}"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        # Directly access the translation text
        translated_text = response.json()["responseData"]["translatedText"]
        return translated_text
    else:
        return "Translation failed"

# Translate to French
french_translation = translate_text("Hello! How are you?", "en", "fr")
print("French Translation:", french_translation)

# Translate to Telugu
telugu_translation = translate_text("Hello! How are you?", "en", "te")
print("Telugu Translation:", telugu_translation)


# Translate to Hindi
hindi_translation = translate_text("Hello! How are you?", "en", "hi")
print("Hindi Translation:", hindi_translation)

```

```python
#@title CSV: Getting CSV through requests module
import requests
import csv
from io import StringIO

# URL of the Titanic CSV dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Fetch the CSV file as plain text
response = requests.get(url)

if response.status_code == 200:
    # Read the CSV text
    csv_text = response.text

    # Use StringIO to read the CSV text into csv.reader
    csv_reader = csv.reader(StringIO(csv_text))

    # Convert to list to view the data as rows
    data = list(csv_reader)

    # Print the header and the first 5 rows
    print("Header:", data[0])
    print("First 5 rows:")
    for row in data[1:]:
        print(row)
else:
    print("Failed to retrieve CSV file")
```

```python
#@title TEXT: Getting plain text
import requests

# URL of a plain text file (in this case, "Pride and Prejudice" by Jane Austen from Project Gutenberg)
url = "https://www.gutenberg.org/files/1342/1342-0.txt"

# Fetch the text data
response = requests.get(url)
if response.status_code == 200:
    # Extract the text content
    text_data = response.text

    # Print the first 1000 characters as a sample
    print(text_data[:10])  # Adjust the slice for more or less content
else:
    print("Failed to retrieve text file")


text_tokens = text_data.split()
print("Total number of tokens: ", len(text_tokens))
print("Unique words in the text: ", len(set(text_tokens)))
```

```python
#@title Getting news feed
# Go ahead and register here for your own KEY https://newsapi.org/
import requests

# Replace with your News API key
API_KEY = 'your_api_key_here'
url = f'https://newsapi.org/v2/top-headlines?country=us&apiKey={API_KEY}'

# Fetch the news data
response = requests.get(url)
if response.status_code == 200:
    news_data = response.json()  # Parse JSON response

    # Check if articles are available
    if news_data.get('status') == 'ok':
        articles = news_data.get('articles', [])

        # Print the headlines
        print("Latest News Headlines:")
        for article in articles:
            title = article.get('title')
            url = article.get('url')
            print(f"- {title} ({url})")
    else:
        print("No articles found.")
else:
    print("Failed to retrieve news data.")
```

```python
#@title Helper function to download a file
# example input = c:/abc/content/lib1.pptx
def downloadFile(a_file_name):
    # extract file name from AFileName
    file_name = a_file_name.split("/")[-1]

    # download image using GET
    raw_image = requests.get(file_name, stream=True)

    print('Downloading ', file_name)
    # save the image recieved into the file
    with open(file_name, 'wb') as fd:
        for chunk in raw_image.iter_content(chunk_size=1024):
            fd.write(chunk)
```
