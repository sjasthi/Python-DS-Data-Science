## ğŸ“š Quick Reference: Merge Types

| Merge Type | Code | What It Does | When to Use |
|------------|------|--------------|-------------|
| **Inner** | `how='inner'` | Only matching rows from both | When you only want records that exist in both DataFrames |
| **Left** | `how='left'` | All from left + matching from right | When you want to keep everything from the main DataFrame |
| **Right** | `how='right'` | All from right + matching from left | When you want to keep everything from the second DataFrame |
| **Outer** | `how='outer'` | Everything from both | When you want to see all records, even without matches |

### Common Parameters:
```python
# When column names are the same
df1.merge(df2, on='ColumnName', how='inner')

# When column names are different
df1.merge(df2, left_on='Col1', right_on='Col2', how='left')

# To see merge source
df1.merge(df2, on='ColumnName', how='outer', indicator=True)
```
