```python
#@title imports
import pandas as pd
import numpy as np
```

```python
#@title sample data
# Create sample data
df = pd.DataFrame({
    'A': np.random.randint(1, 100, 10),
    'B': np.random.randint(1, 100, 10),
    'C': np.random.randint(1, 100, 10),
    'D': np.random.randint(1, 100, 10),
    'E': np.random.randint(1, 100, 10),
}, index=pd.date_range('20250101', periods=10))

# Display the dataframe
df.head(10)
```

```python
#@title visualizations - line plot

# Line Plot
df.plot(kind='line', title='Line Plot')
```

```python
#@title visualizations - bar plot
# Bar Plot (discrete variables)(frequencies)
df.plot(kind='bar', title='Bar Plot')
```

```python
#@title visualizations - Histogram
# Histogram Plot (continuous variables)
df.plot(kind='hist', title='Histogram')
```

```python
#@title visualizations - Box Plot
# Box Plot
df.plot(kind='box', title='Box Plot')
```

```python
#@title visualizations - Area Plot
# Area Plot
df.plot(kind='area', title='Area Plot')
```

```python
#@title visualizations - Scatter Plot
# Scatter Plot(compares two variables) (area of a house; price)
df.plot(kind='scatter', x='A', y='B', title='Scatter Plot')
```

```python
#@title Setting the plot size
#(relies on matplotlib behind the scenes)
# Set the plot size
pd.options.plotting.backend = "matplotlib"
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 6)  # Doubles the default width
```

```python
#@title visualizations (after setting the width)

# Line Plot
df.plot(kind='line', title='Line Plot')

# Bar Plot
df.plot(kind='bar', title='Bar Plot')

# Histogram Plot
df.plot(kind='hist', title='Histogram')

# Box Plot
df.plot(kind='box', title='Box Plot')

# Area Plot
df.plot(kind='area', title='Area Plot')

# Scatter Plot
df.plot(kind='scatter', x='A', y='B', title='Scatter Plot')
```

```python
#@title Using plot methods
# 2. Using df.plot.<plot_type>() Methods
df.plot.line(title='Line Plot')
df.plot.bar(title='Bar Plot')
df.plot.hist(title='Histogram')
df.plot.box(title='Box Plot')
df.plot.area(title='Area Plot')
df.plot.scatter(x='A', y='B', title='Scatter Plot')
```

```python
#@title Series: visualizations
# 3. Using Pandas Series Plotting
df['A'].plot.line(title='Line Plot for Column A')
plt.show()
df['A'].plot.bar(title='Bar Plot for Column A')
plt.show()
df['A'].plot.hist(title='Histogram for Column A')
plt.show()
df['A'].plot.box(title='Box Plot for Column A')
plt.show()
df['A'].plot.area(title='Area Plot for Column A')
plt.show()
```
